```python
import pandas as pd
import matplotlib.pyplot as plt
# Path to the CSV file
csv_file_path = "USA.csv"
df = pd.read_csv(csv_file_path)
```

```python

df.plot(kind='line', x="date", y=["cases","deaths"],logy=True)
plt.ylim([1e2,1e9])

plt.grid(alpha=0.3)
```
